# tokenisation

```python
import tiktoken
```

```python
tokenizer = tiktoken.encoding_for_model("gpt-4o")
```

```python
print(tokenizer.name)
```

```python
tokenizer=tiktoken.get_encoding("o200k_base")
```

```python
prompt = "Vous êtes un expert en softskills. Quelles sont les compétences les plus importantes pour un ingénieur en intelligence artificielle ?"
```

```python
tokens=tokenizer.encode(prompt)
```

```python
print(tokens)
```

```python
print(len(tokens))
```

```python
for token in tokens:
    t=tokenizer.decode_single_token_bytes(token).decode("utf-8")
    print(t,end='|')
```

# Utilisation des llm

```python
from langchain_openai import ChatOpenAI
from dotenv.ipython import load_dotenv
```

```python
load_dotenv(override=True)
```

```python
llm=ChatOpenAI(model="gpt-4o", temperature=0)
```

```python
resp=llm.invoke(input=[{"role":"system","content":"Vous êtes un expert en softskills"},{"role":"user","content":"Quelles sont les compétences les plus importantes pour un ingénieur en intelligence artificielle ?"}])
```

```python
print(resp.content)
```

```python
from IPython.display import Markdown
```

```python
print(display(Markdown(resp.content)))
```
